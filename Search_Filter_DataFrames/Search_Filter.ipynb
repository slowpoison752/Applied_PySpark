```python
import findspark
findspark.init()

import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('Search').getOrCreate()
spark
```

```python
path = 'Datasets/'
fifa = spark.read.csv(path + 'fifa19.csv', inferSchema=True, header=True)
fifa.show(5)
```

```python
fifa.limit(5).toPandas()
```

```python
fifa.printSchema()
```

```python
from pyspark.sql.functions import *
```

```python
fifa_trim = fifa.select('Name','Nationality','Age')
```

```python
fifa_trim.orderBy(fifa['Age']).show(5)
```

```python
fifa_trim.orderBy(fifa['Age'].desc()).show(5)
```

```python
fifa.select(['Name','Club']).where(fifa.Club.like("%Barcelona%")).show(5,False)
```

```python
fifa[fifa.Club.isin("FC Barcelona","Juventus")].limit(4).toPandas()
```

```python
fifa.select(['Name','Club']).where(fifa.Club.isin("FC Barcelona","Juventus")).limit(5).toPandas()
```

```python
fifa.select(['Name','Club']).where(fifa.Name.startswith("A")).where(fifa.Name.endswith("a")).limit(5).toPandas()
```

## Filter

```python
fifa.filter('Overall>85').limit(5).toPandas()
```

```python
fifa.filter('Overall<90').filter('Overall>85').limit(5).toPandas()
```

```python
fifa.filter('Overall<90').filter('Overall>85').select('Name','Age').limit(5).toPandas()
```

```python
result = fifa.filter('Overall<90').filter('Overall>85').select('Name','Age','Nationality','Overall').orderBy(fifa['Overall'].desc()).collect()
```

```python
type(result)
```

```python
print ("Best Player below 90:",result[0][0])
```

```python
print ("Worst Player below 90 and above 85:",result[-1][0])
```
